```python
import pandas as pd
import sqlite3
conn = sqlite3.connect("dbConnect")
df = pd.read_csv("database.csv")
df.to_sql("resource", conn, if_exists="append",index=False)
```

```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Exception as e:
        print(e)
    return conn

def select_all(conn):
    cur = conn.cursor()
    cur.execute(f"select * from resource")
    rows = cur.fetchall()
    print(rows)
    
```

```python
select_all(create_connection("dbConnect"))
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)

def select_answers(conn, question):
    rows = []
    cur = conn.cursor()
    cur.execute(f"select answers from resource where questions like '%{str(question.replace(' ', '%'))}%'")
    rows = cur.fetchall()
    if rows == []:
        for q in question.split():
            cur.execute(f"select distinct answers from resource where questions like '%{q}%'")
            rows.append(cur.fetchall())
    flat = [x for sublist in rows for x in sublist]
    print(most_common(flat)[0])
    
select_answers(create_connection("dbConnect"), 'ഫ്രീഷിപ് കാർഡ്')
```
